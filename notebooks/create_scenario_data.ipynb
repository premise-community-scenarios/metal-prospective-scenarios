```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
#Import data
excel_file = "/Users/margueritefauroux/Desktop/data_metals.xlsx"
sheet_name_cu_prod = "cu_prod"
df_cu_prod = pd.read_excel(excel_file, sheet_name=sheet_name_cu_prod)

sheet_name_cu_dem = "cu_dem"
df_cu_dem = pd.read_excel(excel_file, sheet_name=sheet_name_cu_dem)
period = [2022, 2025, 2030, 2035, 2040, 2045, 2050]
df_cu_dem = df_cu_dem[df_cu_dem['Year'].isin(period)].set_index('Year')
```

```python
#Share of copper production locations in 2019
#We will make the hypothesis that these shares stay constant
#Refining
total_cu_prod_ref = df_cu_prod['Prod Refinery Primary tCu'].sum()
share_cu_prod_ref_by_region = (df_cu_prod.groupby('Region')['Prod Refinery Primary tCu'].sum() / total_cu_prod_ref)
print(share_cu_prod_ref_by_region.sort_values(ascending=False))
```

```python
#Leaching
total_cu_prod_leach = df_cu_prod['Prod Leach tCu'].sum()
share_cu_prod_leach_by_region = (df_cu_prod.groupby('Region')['Prod Leach tCu'].sum() / total_cu_prod_leach)
print(share_cu_prod_leach_by_region.sort_values(ascending=False))
```

```python
#Secondary production
total_cu_prod_sec = df_cu_prod['Prod Refinery Secondary tCu'].sum()
share_cu_prod_sec_by_region = (df_cu_prod.groupby('Region')['Prod Refinery Secondary tCu'].sum() / total_cu_prod_sec)
print(share_cu_prod_sec_by_region.sort_values(ascending=False))
```

```python
#Shares of copper primary production pathways (refining and leaching) in 2019
#We will make the hypothesis that these shares stay constant
share_ref = df_cu_prod['Prod Refinery Primary tCu'].sum() / (df_cu_prod['Prod Refinery Primary tCu'].sum() + df_cu_prod['Prod Leach tCu'].sum())
share_leach = df_cu_prod['Prod Leach tCu'].sum() / (df_cu_prod['Prod Refinery Primary tCu'].sum() + df_cu_prod['Prod Leach tCu'].sum())
print(share_ref)
print(share_leach)
```

```python
#Quantity of copper supplied by refining for each year and each scenario
prod_ref_NZE = df_cu_dem['primary production']*df_cu_dem['NZE']*share_ref
prod_ref_SPS = df_cu_dem['primary production']*df_cu_dem['SPS']*share_ref
```

```python
#Quantity of copper supplied by leaching for each year and each scenario
prod_leach_NZE = df_cu_dem['primary production']*df_cu_dem['NZE']*share_leach
prod_leach_SPS = df_cu_dem['primary production']*df_cu_dem['SPS']*share_leach
```

```python
#Quantity of copper supplied by secondary production for each year and each scenario
prod_sec_NZE = df_cu_dem['secondary production']*df_cu_dem['NZE']
prod_sec_SPS = df_cu_dem['secondary production']*df_cu_dem['SPS']
```

```python
#Quantity of copper produced for each production pathway and for each scenario
#The dataframe created contains the quantities for each year and for each region
prod_ref_NZE_by_region = pd.DataFrame(columns=prod_ref_NZE.index, index=share_cu_prod_ref_by_region.index)
for col in prod_ref_NZE_by_region.columns:
    prod_ref_NZE_by_region[col] = prod_ref_NZE[col] * share_cu_prod_ref_by_region

prod_leach_NZE_by_region = pd.DataFrame(columns=prod_leach_NZE.index, index=share_cu_prod_leach_by_region.index)
for col in prod_leach_NZE_by_region.columns:
    prod_leach_NZE_by_region[col] = prod_leach_NZE[col] * share_cu_prod_leach_by_region

prod_sec_NZE_by_region = pd.DataFrame(columns=prod_sec_NZE.index, index=share_cu_prod_sec_by_region.index)
for col in prod_sec_NZE_by_region.columns:
    prod_sec_NZE_by_region[col] = prod_sec_NZE[col] * share_cu_prod_sec_by_region

prod_ref_SPS_by_region = pd.DataFrame(columns=prod_ref_SPS.index, index=share_cu_prod_ref_by_region.index)
for col in prod_ref_SPS_by_region.columns:
    prod_ref_SPS_by_region[col] = prod_ref_SPS[col] * share_cu_prod_ref_by_region

prod_leach_SPS_by_region = pd.DataFrame(columns=prod_leach_SPS.index, index=share_cu_prod_leach_by_region.index)
for col in prod_leach_SPS_by_region.columns:
    prod_leach_SPS_by_region[col] = prod_leach_SPS[col] * share_cu_prod_leach_by_region

prod_sec_SPS_by_region = pd.DataFrame(columns=prod_sec_SPS.index, index=share_cu_prod_sec_by_region.index)
for col in prod_sec_SPS_by_region.columns:
    prod_sec_SPS_by_region[col] = prod_sec_SPS[col] * share_cu_prod_sec_by_region
```

```python
data_NZE = []
for index, row in prod_ref_NZE_by_region.iterrows():
    remind = "remind"
    SSP2_PkBudg500 = "SSP2-PkBudg500"
    NZE = "NZE"
    region = index
    production_type = "Production|Copper|Electrorefining"
    unit = "ton"
    row_rounded = [round(val, 2) for val in row]
    data_NZE.append([remind, SSP2_PkBudg500, NZE, region, production_type, unit] + row_rounded)
```
